```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# ---------------------------
# Task 1: Dataset Creation
# ---------------------------
data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Math": [85, 72, 90, 60, 95],
    "Science": [78, 65, 88, 75, 92],
    "English": [90, 70, 95, 68, 96]
}

df = pd.DataFrame(data)
print("Dataset:\n", df)
```

```python
#use the numpy to calculate average marks of each student
marks_array=df[["Math","Science","English"]].to_numpy()
student_avg=np.mean(marks_array,axis=1)
df["Average"]=student_avg
print("Average marks of each student:\n",df[["Student","Average"]])
```

```python
#subject average
# sub_avg=np.mean(marks_array,axis=0) OR 
subject_avg = df[["Math", "Science", "English"]].mean()
print("Average marks of each subject:\n",subject_avg)
```

```python
#student with highest toatal marks
df["Total"] =df[["Math","Science","English"]].sum(axis=1)
top_student=df.loc[df["Total"].idxmax()]
print("Student with highest total marks:",top_student)
```

```python
df["Result"]=np.where(df["Average"]>=70,"Pass","Fail")
print("Result of each student:\n",df[["Student","Result"]])
```

```python
df["Result"] = np.where(df["Average"] >= 70, "Pass", "Fail")
print("\nFinal DataFrame with Result:\n", df)
```

```python
# (a) Bar chart of each students total marks
plt.bar(df["Student"], df["Total"], color="skyblue")
plt.title("Total Marks of Students")
plt.xlabel("Student")
plt.ylabel("Total Marks")
plt.show()

# (b) Line chart of subject-wise average marks
plt.plot(subject_avg.index, subject_avg.values, marker="o", color="green")
plt.title("Subject-wise Average Marks")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.grid(True)
plt.show()

# (c) Scatter plot of Math vs Science marks
plt.scatter(df["Math"], df["Science"], color="red")
plt.title("Math vs Science Marks")
plt.xlabel("Math Marks")
plt.ylabel("Science Marks")
plt.show()
```
