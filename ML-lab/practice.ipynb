```python
import numpy as np

# 1. Creating arrays
a = np.array([1, 2, 3, 4, 5])
print("1D Array:", a)

b = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", b)

# 2. Array of zeros and ones
zeros = np.zeros((3, 3))
ones = np.ones((2, 4))
print("\nZeros:\n", zeros)
print("\nOnes:\n", ones)

# 3. Range and linspace
arr_range = np.arange(0, 10, 2)
arr_lin = np.linspace(0, 1, 5)
print("\nRange Array:", arr_range)
print("Linspace Array:", arr_lin)

# 4. Array operations
x = np.array([10, 20, 30, 40])
y = np.array([1, 2, 3, 4])
print("\nAddition:", x + y)
print("Multiplication:", x * y)
print("Division:", x / y)

# 5. Matrix operations
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7, 8]])
print("\nMatrix Multiplication:\n", np.dot(mat1, mat2))

# 6. Statistical functions
random_arr = np.random.randint(1, 100, 10)
print("\nRandom Array:", random_arr)
print("Mean:", np.mean(random_arr))
print("Standard Deviation:", np.std(random_arr))
print("Max:", np.max(random_arr))
print("Min:", np.min(random_arr))

# 7. Reshaping
reshaped = np.arange(1, 13).reshape(3, 4)
print("\nReshaped Array (3x4):\n", reshaped)
```

```python
import pandas as pd

# 1. Creating a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 22],
        'City': ['New York', 'San Francisco', 'Los Angeles']}
df = pd.DataFrame(data)
print(df)

# 2. Reading data from a CSV file
# # df = pd.read_csv('data.csv')

# 3. Basic operations
print(df.head())  # First few rows
print(df.tail())  # Last few rows
print(df.info())  # Summary of DataFrame
print(df.describe())  # Statistical summary

# 4. Indexing and Slicing
print(df['Name'])  # Column
print(df.iloc[0])  # Row
print(df.loc[0])  # Row
print(df.iloc[1:3])  # Rows 1 to 2

```

```python
import pandas as pd

# Create a DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [24, 30, 22],
    "City": ["New York", "London", "Paris"]
}

df = pd.DataFrame(data)

print("DataFrame:\n", df)

# Access columns
print("Names:", df["Name"])

# Filtering
print("Age > 23:\n", df[df["Age"] > 23])

# Summary statistics
print("Statistics:\n", df.describe())
```

```python
import matplotlib.pyplot as plt

# 1. Line plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# 2. Scatter plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# 3. Bar plot
x = ["A", "B", "C", "D", "E"]
y = [10, 15, 13, 17, 20]
plt.bar(x, y)
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()

# 4. Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=5)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

sizes = [30, 20, 25, 25]
labels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie Chart")
plt.show()
```
